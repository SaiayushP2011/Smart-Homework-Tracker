```python
assignments = [] #list for keeping track of all assignments
#basic menu for homework tracker
print("---- Homework Tracker----")
print("1. Add assignment")
print("2. View assignments")
print("3. Complete assignment")
print("4. Save and Exit")
#loop to keep the program running until the user decides to choose the 4th option to save and exit
while(True):
    try:
        #sorting system based on priority and due date, with high priority and sooner due dates appearing first
        priority_order = {"High": 3, "Medium": 2, "Low": 1}
        #sorts the assignments list based on the priority and due date, with higher priority and sooner due dates appearing first
        assignments.sort(key=lambda x: (-priority_order.get(x[1], 0), int(x[2])))
        choice = int(input("Enter your choice: "))
        if choice == 1:
            #takes in the assignment name, priority, and due date and adds it to the list of assignments
            assignment = input("Enter the assignment name: ")
            priority = input("Enter the priority (High/Medium/Low): ")
            due_date = input("Enter the number of days until due: ")
            assignments.append((assignment, priority, due_date))
            print(f"Assignment '{assignment}' added.")
        elif choice == 2:
            print("Viewing assignments...")
            for i in range(len(assignments)):
                print(f"{assignments[i][0]} - Priority: {assignments[i][1]}, Due in: {assignments[i][2]} days")
        elif choice == 3:
            assignment = input("Enter the assignment name to complete: ")
            #checks for the assignment in the list and removes it if found, otherwise it prints that the assignment was not found
            for i in range(len(assignments)):
                if assignments[i][0] == assignment:
                    assignments.pop(i)
                    print(f"Assignment '{assignment}' marked as completed.")
                    break
            else:
                print(f"Assignment '{assignment}' not found.")
        elif choice == 4:
            print("Saving and exiting...")
            break
        else:
            print("Invalid choice. Please try again.")
    except ValueError:
        print("Please enter a valid numerical value between 1 and 4.")
```
